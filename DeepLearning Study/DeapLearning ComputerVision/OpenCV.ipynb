## 이미지로딩

* imread()를 이용한 이미지로딩을 수행하면 RGB가 아닌 BGR형태로 로딩을 한다
 * cvtColor()를 이용해 RGB로 바꾼다


```python
import cv2
import matplotlib.pyplot as plt
img_array = cv2.imread("파일명") # 이미지 로딩
rgb_img_array = cv2.cvtColor(img_array,cv2.COLOR_BGR2RGB) # BGR to RGB
plt.imshow(rgb_img_array)

cv2.imwrite("파일명",img_array) #파일 쓰기 --> BGR to RGB를 해줌
cv2.waitkey() # 키보드 입력이 있을 때 까지 무한 대기
cv2.destroyAllWindows() # 화면의 윈도우 프레임 모두 종료
```

## Object Detection 네트웍 개요

* Feature Extractor Network - 백본이라고도 불림 --> 원본이미지에서 주요 Feature들을 뽑아내는 역할
 * imageNet 데이터 세트 기반으로 PreTrained 됨
 
* Object Detection Network - Bounding Box , Classification 등등 결정
* Region Proposal - Object가 있을만한 곳을 결정



* 높은 Image Resolution - 선명한 화질 (Detection 성능 향상, FPS(frame per second)가 떨어짐 --> 배열이 많이 큼)
* FPS가 빨라야 하는 요구사항 - Detection 성능 하락

--> FPS, Detection 성능 둘 다 높은 알고리즘은 없다!
