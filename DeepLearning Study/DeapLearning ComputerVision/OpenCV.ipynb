* imread()    RGB  BGR

```python
import cv2
import matplotlib.pyplot as plt
img_array = cv2.imread("") #  
rgb_img_array = cv2.cvtColor(img_array,cv2.COLOR_BGR2RGB) # BGR to RGB
plt.imshow(rgb_img_array)

cv2.imwrite("",img_array) #  --> BGR to RGB 
cv2.waitkey() #       
cv2.destroyAllWindows() #     
```
