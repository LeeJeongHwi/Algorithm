# SSD
## Single Shot (Mutibox) Detector)

**RCNN 계열은 Two Stage Detector 이다**
* 가장 빠른 Faster RCNN 7FPS 라서 실시간으로는 사용 불가했다.

* YOLO는 Faster R-CNN 보다 빠르긴 했지만, 성능이 좋지 않았다

#### SSD
* Faster R-CNN 보다 빠르고, YOLO보다 성능이 좋음
* Anchor Box에 기반한 Object Detection 모델
> Multi Scale Feature Layer + Default(Anchor) Box

### Multi Scale Feature Layer

**서로 다른 크기의 Feature Map들을 기반으로 Object Detection**

* Feature Map이 작아지면서 추상적인 이미지의 형태로 변경 --> 핵심 적인 영역만 갖게 됨


* Feature Map 의 크기가 작을 수록 더 큰 이미지를 찾을 수 있다.
* 개별 Feature Map에 대해서 Detection을 해주고 각각 Detect 한 결과를 가지고 합쳐서 최종적인 Object Detection을 수행

### Anchor Box 기반의 Object Detection 모델

* 개별 Anchor box 별로 Detection 하려는 Object 유형 Softmax 값, 수정 좌표 값
