```python
import pandas as pd
import matplotlib.pyplot as plt
marathon_2015_2017 = pd.read_csv("./data/marathon_2015_2017.csv")
```

```python
record = pd.DataFrame(marathon_2015_2017,columns=['5K','10K','15K','20K','Half','25K','30K','35K','40K',"Official Time"])
```

```python
record.insert(0,'Rank',range(1,1+len(record)))
top100 = record[0:101]
xData = top100.Rank

yData_full = top100['Official Time']
yData_10K = top100['10K']
yData_20K = top100['20K']
yData_30K = top100['30K']
```

```python
plt.figure(figsize=(20,10))

#color,linewidth,linestyle,marker,label
plt.plot(xData,yData_full,color='blue',linewidth=4,linestyle='-',marker='x',label='Full Time')
plt.plot(xData,yData_10K,color='red',linewidth=3,linestyle='--',marker='s',label='10K')
plt.plot(xData,yData_20K,color='orange',linewidth=2,linestyle='-.',marker='o',label='20K')
plt.plot(xData,yData_30K,color='green',linewidth=1,linestyle=':',marker='d',label='30K')
#title
plt.title("Top 100 Ranker's Records")

#x,y label name
plt.xlabel("Rank")
plt.ylabel("Time(second)")

#Legend (label )
plt.legend() 

plt.show()
```
