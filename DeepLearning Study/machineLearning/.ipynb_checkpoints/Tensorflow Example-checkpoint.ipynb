```python
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior()
```

```python
#상수는 Constant
a = tf.constant(2,tf.float32)
b = tf.constant(3,tf.float32)
c = tf.add(a,b)
sess = tf.Session()
sess.run(c)
```

```python
#Variable
a = tf.Variable(2)
b = tf.Variable(3)
c = tf.add(a,b)
sess = tf.Session()
initial = tf.global_variables_initializer() # 변수초기화
sess.run(initial)
sess.run(c)
#placeHolder
#tf.placeholder(dtype,shape=[],name="")
```

```python
import tensorflow as tf
import numpy as np
```

```python
#TensorFlow2 coding Style
#tensorflow2는 기본적으로 keras를 기본으로한다
x_train = [1,2,3,4]
y_train = [1,2,3,4]

model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Dense(1,input_dim=1))#dense(output,input_dimension)

#optimizer
sgd = tf.keras.optimizers.SGD(lr=0.01)
#SGD : stochastic gradient Descent

#cost/loss function
model.compile(loss='mean_squared_error',optimizer=sgd)

#fit the line
model.fit(x_train,y_train,epochs = 2000)

print(model.predict(np.array([5])))
```
